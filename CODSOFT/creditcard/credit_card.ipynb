```python
# Import Libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load Dataset
file = pd.read_csv("creditcard.csv")

# Basic Info
print(file.head())
print(file.describe())
print(file.isnull().sum())

# Check Class Distribution
print(file['Class'].value_counts())

# Separate normal and fraud
normal = file[file.Class == 0]
fraud = file[file.Class == 1]

print("Normal shape:", normal.shape)
print("Fraud shape:", fraud.shape)

# Compare Amount stats
print("Normal Amount stats:\n", normal.Amount.describe())
print("Fraud Amount stats:\n", fraud.Amount.describe())

# Group by class
print(file.groupby('Class').mean())

# Handle Imbalance (Undersampling)
normal_sample = normal.sample(n=492, random_state=2)

# Combine datasets
new_file = pd.concat([normal_sample, fraud], axis=0)

# Shuffle dataset
new_file = new_file.sample(frac=1, random_state=2)

print(new_file.head())
print(new_file['Class'].value_counts())

# Split Features & Target
X = new_file.drop(columns='Class')
Y = new_file['Class']

# Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, stratify=Y, random_state=2
)

# Model Training
model = LogisticRegression(max_iter=5000)
model.fit(X_train, Y_train)

# Training Accuracy
X_train_pred = model.predict(X_train)
train_acc = accuracy_score(Y_train, X_train_pred) * 100
print(f"Training Accuracy: {train_acc:.2f}%")

# Testing Accuracy
X_test_pred = model.predict(X_test)
test_acc = accuracy_score(Y_test, X_test_pred) * 100
print(f"Test Accuracy: {test_acc:.2f}%")

# Final Check
print(new_file.groupby('Class').mean())
```
